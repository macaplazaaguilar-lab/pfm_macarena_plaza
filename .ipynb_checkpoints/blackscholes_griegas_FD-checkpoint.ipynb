```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
#Calcula el precio de una opción europea (call o put) usando Black-Scholes.
   # CP: "call" o "put"
   # S_0: precio inicial del activo
   # K: strike
   # sigma: volatilidad
   # t: tiempo actual
   # T: vencimiento
   # r: tipo de interés libre de riesgo

def BS_Call_Put_Option_Price(CP,S_0,K,sigma,t,T,r):
    if T - t <= 0:
        # Opción vencida
        if CP == "call":
            return max(S_0 - K, 0)
        else:
            return max(K - S_0, 0)

    K = float(K)#np.array(K).reshape([len(K),1])
    d1    = (np.log(S_0 / K) + (r + 0.5 * np.power(sigma,2.0)) * (T-t)) / (sigma * np.sqrt(T-t))
    d2    = d1 - sigma * np.sqrt(T-t)
    
    if CP.lower()  == "call":
        value = norm.cdf(d1) * S_0 - norm.cdf(d2) * K * np.exp(-r * (T-t))
    elif CP.lower()  == "put":
        value = norm.cdf(-d2) * K * np.exp(-r * (T-t)) - norm.cdf(-d1)*S_0

    return value
```

### Griegas

#### 1. DELTA 

```python
#DELTA, valor analítico
def BS_delta(CP, S0, K, sigma, t, T, r): 
    tau = T - t
    if tau <= 0:
        # En vencimiento
        if CP.lower() == "call":
            return 1.0 if S0 > K else 0.0
        elif CP.lower() == "put":
            return -1.0 if S0 < K else 0.0
    
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * tau) / (sigma * np.sqrt(tau))
    
    if CP.lower() == "call":
        return norm.cdf(d1)
    elif CP.lower() == "put":
        return norm.cdf(d1) - 1
    else:
        raise ValueError("CP debe ser 'call' o 'put'")

#DELTA, con diferencias finitas (centrada)
def BS_delta_FD(CP, S0, K, sigma, t, T, r,h):
    
    price_up = BS_Call_Put_Option_Price(CP, S0 + h, K, sigma, t, T, r)
    price_down = BS_Call_Put_Option_Price(CP, S0 - h, K, sigma, t, T, r)
    
    delta = (price_up - price_down) / (2*h)
    
    return delta
    
```

```python
#test
S0 = 100
K = 100
sigma = 0.2
r = 0.05
t = 0
T = 1

h = 0.01 * S0

delta_exact = BS_delta("call", S0, K, sigma, t, T, r)
delta_fd = BS_delta_FD("call", S0, K, sigma, t, T, r, h)

print("Delta exacta:", delta_exact)
print("Delta FD:", delta_fd)
print("Error absoluto:", abs(delta_exact - delta_fd))
```

```python
#vamos a recorrerlo para diferentes valores de h
hs = np.logspace(-6, 0, 60) #Genera 60 valores de h distribuidos logarítmicamente entre: 10^-6 y 10^0

errors = []

for h in hs:
    delta_fd = BS_delta_FD("call", S0, K, sigma, t, T, r, h)
    error = abs(delta_exact - delta_fd)
    errors.append(error)

errors = np.array(errors)

#representamos en gráfica
plt.figure(figsize=(8,6))
plt.loglog(hs, errors, 'o-', label="Error FD")

# Pendiente de referencia con va slope 2 line
plt.loglog(hs, hs**2 * errors[30]/hs[30]**2, '--', label="Recta con pendiente 2")


plt.xlabel("h")
plt.ylabel("Error Absoluto")
plt.title("Gráfica Log-Log Error Delta (Central FD)")
plt.legend()
plt.grid(True, which="both")
plt.show()
```

La gráfica log-log muestra una forma en U, que creo que es por la combinación de dos tipos de error: el error de truncamiento, cuando 
h es grande, y el error de redondeo (porque los valores de la opción son muy similares en S+h y S-h), cuando h es muy pequeño.
En la zona intermedia se observa que la pendiente es aproximadamente 2, lo que indica que el método de diferencias finitas centradas tiene convergencia de orden dos.

#### 2. GAMMA

Gamma es la segunda derivada respecto a S.\
Como Call y Put solo difieren en términos lineales en S.\
Al derivar dos veces, esa diferencia desaparece.

```python
#GAMMA, valor analítico
def BS_gamma(S0, K, sigma, t, T, r):
    tau = T - t
    
    if tau <= 0:
        return 0.0
    
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * tau) / (sigma * np.sqrt(tau))
    
    gamma = norm.pdf(d1) / (S0 * sigma * np.sqrt(tau))
    
    return gamma

#GAMMA, por diferencias finitas
def BS_gamma_FD(CP, S0, K, sigma, t, T, r, h):
    
    price_up = BS_Call_Put_Option_Price(CP, S0 + h, K, sigma, t, T, r)
    price_mid = BS_Call_Put_Option_Price(CP, S0, K, sigma, t, T, r)
    price_down = BS_Call_Put_Option_Price(CP, S0 - h, K, sigma, t, T, r)
    
    gamma = (price_up - 2*price_mid + price_down) / (h**2)
    
    return gamma
```

```python
#test
S0 = 100
K = 100
sigma = 0.2
r = 0.05
t = 0
T = 1

h = 0.01 * S0

gamma_exact = BS_gamma(S0, K, sigma, t, T, r)
gamma_fd = BS_gamma_FD('call',S0, K, sigma, t, T, r, h)

print("Gamma exacta:", gamma_exact)
print("Gamma FD:", gamma_fd)
print("Error absoluto:", abs(gamma_exact - gamma_fd))
```

```python
# Vamos a recorrerlo para diferentes valores de h
hs = np.logspace(-6, 0, 60)  # 60 valores logarítmicos entre 1e-6 y 1

errors = []

for h in hs:
    gamma_fd = BS_gamma_FD("call", S0, K, sigma, t, T, r, h)  # Call o Put da igual
    error = abs(gamma_exact - gamma_fd)
    errors.append(error)

errors = np.array(errors)

# Representamos en gráfica log-log
plt.figure(figsize=(8,6))
plt.loglog(hs, errors, 'o-', label="Error FD")

# Línea de referencia con pendiente 2
plt.loglog(hs, hs**2 * errors[30]/hs[30]**2, '--', label="Recta pendiente 2")

plt.xlabel("h")
plt.ylabel("Error Absoluto")
plt.title("Gráfica Log-Log Error Gamma (Central FD)")
plt.legend()
plt.grid(True, which="both")
plt.show()
```

* Para valores grandes de h (a la derecha): el error es grande porque la aproximación por diferencias finitas es menos precisa, ya que el paso es muy grande (error de truncamiento).

* Para valores muy pequeños de h (a la izquierda): el error también sube porque aparecen errores numéricos por redondeo en los cálculos con números muy pequeños.

* En la zona intermedia: se aprecia una pendiente que sigue muy bien la línea naranja de pendiente 2, lo que indica que el método de diferencias finitas centrales para Gamma tiene un error que decrece aproximadamente como ℎ^2. Esto confirma que la fórmula tiene orden 2 de precisión, justo lo esperado.


#### 3. VEGA

La Vega mide la sensibilidad del precio respecto a la volatilidad:

```python
# VEGA, valor analítico
def BS_vega(S0, K, sigma, t, T, r):
    
    tau = T - t
    
    if tau <= 0:
        return 0.0
    
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * tau) / (sigma * np.sqrt(tau))
    
    vega = S0 * norm.pdf(d1) * np.sqrt(tau)
    
    return vega

#VEGA, con diferencias finitas
def BS_vega_FD(CP, S0, K, sigma, t, T, r, h):
    
    price_up = BS_Call_Put_Option_Price(CP, S0, K, sigma + h, t, T, r)
    price_down = BS_Call_Put_Option_Price(CP, S0, K, sigma - h, t, T, r)
    
    vega = (price_up - price_down) / (2*h)
    
    return vega
```

```python
# Test VEGA
S0 = 100
K = 100
sigma = 0.2
r = 0.05
t = 0
T = 1

h = 0.01 * sigma  

vega_exact = BS_vega(S0, K, sigma, t, T, r)
vega_fd = BS_vega_FD('call', S0, K, sigma, t, T, r, h)

print("Vega exacta:", vega_exact)
print("Vega FD:", vega_fd)
print("Error absoluto:", abs(vega_exact - vega_fd))
```

```python
# Vamos a recorrerlo para diferentes valores de h
hs = np.logspace(-8, -1, 60)  # pasos pequeños sobre sigma

errors = []

for h in hs:
    vega_fd = BS_vega_FD("call", S0, K, sigma, t, T, r, h)
    error = abs(vega_exact - vega_fd)
    errors.append(error)

errors = np.array(errors)

# Representamos en gráfica log-log
plt.figure(figsize=(8,6))
plt.loglog(hs, errors, 'o-', label="Error FD")

# Línea de referencia con pendiente 2 (orden 2 esperado)
plt.loglog(hs, hs**2 * errors[30]/hs[30]**2, '--', label="Recta pendiente 2")

plt.xlabel("h")
plt.ylabel("Error Absoluto")
plt.title("Gráfica Log-Log Error Vega (Central FD)")
plt.legend()
plt.grid(True, which="both")
plt.show()
```

- Zona derecha — h grande. Vemos que la pendiente es más o menos 2. Aquí domina el error de truncamiento.

    Como usamos diferencias centrales: El error teórico es: Error=O(h^2).
    En log-log eso es una recta de pendiente 2.

- Zona izquierda — h muy pequeño. Aquí el error deja de bajar y se aplana. Esto es error de redondeo (floating point).

    Cuando h es muy pequeño: restamos números casi iguales entonces se pierde precisión. 

- Mínimo del error (zona óptima). Vemos que el mínimo está aproximadamente en: h≈10−6. Ahí el error baja hasta ≈ 10⁻¹⁰.

    Ese es el punto donde: error truncamiento≈error redondeo


#### 4. RHO

Rho mide la sensibilidad respecto al tipo libre de riesgo

```python
# RHO, valor analítico
def BS_rho(CP, S0, K, sigma, t, T, r):
    
    tau = T - t
    
    if tau <= 0:
        return 0.0
    
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * tau) / (sigma * np.sqrt(tau))
    d2 = d1 - sigma * np.sqrt(tau)
    
    if CP.lower() == "call":
        rho = K * tau * np.exp(-r * tau) * norm.cdf(d2)
    else:
        rho = -K * tau * np.exp(-r * tau) * norm.cdf(-d2)
    
    return rho
```

```python
# RHO, por diferencias finitas
def BS_rho_FD(CP, S0, K, sigma, t, T, r, h):
    
    price_up = BS_Call_Put_Option_Price(CP, S0, K, sigma, t, T, r + h)
    price_down = BS_Call_Put_Option_Price(CP, S0, K, sigma, t, T, r - h)
    
    rho = (price_up - price_down) / (2*h)
    
    return rho
```

```python
# TEST RHO
S0 = 100
K = 100
sigma = 0.2
r = 0.05
t = 0
T = 1

h = 0.01 * r   # paso relativo sobre r

rho_exact = BS_rho("call", S0, K, sigma, t, T, r)
rho_fd = BS_rho_FD("call", S0, K, sigma, t, T, r, h)

print("Rho exacta:", rho_exact)
print("Rho FD:", rho_fd)
print("Error absoluto:", abs(rho_exact - rho_fd))
```

```python
hs = np.logspace(-8, -1, 60)

errors = []

for h in hs:
    rho_fd = BS_rho_FD("call", S0, K, sigma, t, T, r, h)
    error = abs(rho_exact - rho_fd)
    errors.append(error)

errors = np.array(errors)

plt.figure(figsize=(8,6))
plt.loglog(hs, errors, 'o-', label="Error FD")

# Recta pendiente 2
plt.loglog(hs, hs**2 * errors[30]/hs[30]**2, '--', label="Recta pendiente 2")

plt.xlabel("h")
plt.ylabel("Error Absoluto")
plt.title("Gráfica Log-Log Error Rho (Central FD)")
plt.legend()
plt.grid(True, which="both")
plt.show()
```

El análisis del error es similar al de Vega.

Rho es mayor cuando hay más tiempo al vencimiento (proporcional a τ).
Es pequeña para opciones cortas.

- Calls tienen Rho positiva.
- Puts tienen Rho negativa.

Si suben los tipos → aumenta el valor presente del subyacente relativo al strike → favorece calls.
