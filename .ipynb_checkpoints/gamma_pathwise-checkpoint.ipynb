Bajo BlackScholes:

_T=_0*exp((r-1/2^2)+_T)
S_T/S_0=G

Entonces: La derivada de S_T respecto a S_0 es G pero la segunda derivada es 0.\
Y lo mismo pasa co el payoff, la segunda derivada respecto a S_T es 0

- Por lo tanto el m茅todo de pathwise para gamma falla.
- El payoff no es dos veces diferenciable en 
- Gamma depende de la masa en el strike
- Pathwise no captura la contribuci贸n tipo delta de Dirac

```python
#El c贸digo ser铆a as铆, pero nos dar铆a 0 todo el rato. Es una limitaci贸n de pathwise
```

```python
def Gamma_pathwise_call(S0, K, r, sigma, T, Z):
    
    # 1) Simular S_T
    ST = S0 * np.exp((r - 0.5*sigma**2)*T + sigma*np.sqrt(T)*Z)
    
    # 2) Indicador ITM
    indicator = (ST > K).astype(float)
    
    # 3) Derivadas del path
    dST_dS0 = ST / S0
    d2ST_dS02 = 0.0   # <- clave
    
    # 4) Derivadas payoff
    f_prime = indicator
    f_second = 0.0    # <- casi seguro
    
    # 5) F贸rmula general Gamma
    gamma = np.exp(-r*T) * np.mean(
        f_second * dST_dS0**2 +
        f_prime * d2ST_dS02
    )
    
    return gamma
```
